# 14 Bracket Placement
### Wrong Bracket placement
- we wanted that only femlaes from both towns should be there not males, but it is also showing male as well

```python
df[((df.embark_town=='Southampton') |
   (df.embark_town=='Queenstown')  &
    (df.sex=='female') )]
```

### correct bracket placement

```python
df[((df.embark_town=='Southampton') |
   (df.embark_town=='Queenstown'))  &
    (df.sex=='female')]
```

```python
# jana par sirf Queenstown ka log ho
df[df.embark_town.isin(['Queenstown'])].head()
```

- where age > 30

```python
df[((df.embark_town=='Southampton') |
   (df.embark_town=='Queenstown'))  &
    (df.age>30)]
```

```python
# assignment age more than 30
df[df.age>30]
```

## 16- splitting a string into multiple coulumns

```python
#import libraries
import pandas as pd
df = pd.DataFrame({'name':['Abdul raheem','sajad ali','subman gil'],
                   'location':['Peshawar, Pakistan','lahore, Pakistan','mumbai, india'],})
df
```

```python
df.name.str.split(' ', expand = True)
```
