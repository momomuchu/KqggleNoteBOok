Anomaly Detection means to identify unexpected events in a process. It means to detect threats to our systems that may cause harm in terms of security and leakage of important information.


 
The importance of Anomaly Detection is not limited to security, but it is used for detection of any event that does not conform to our expectations.

```python
import pandas as pd
import pyflux as pf
from datetime import datetime
```

```python
data_train_a = pd.read_csv('../input/cpu-utilization/cpu-train-a.csv', parse_dates=[0], infer_datetime_format=True)
data_test_a = pd.read_csv('../input/cpu-utilization/cpu-test-a.csv', parse_dates=[0], infer_datetime_format=True)
data_train_a.head()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,8))
plt.plot(data_train_a['datetime'], data_train_a['cpu'], color='black')
plt.ylabel('CPU %')
plt.grid()
plt.title('CPU Utilization')
plt.show()
```

```python
model_a = pf.ARIMA(data=data_train_a, ar=11, ma=11, integ=0, target='cpu')
x = model_a.fit("M-H")
```

```python
model_a.plot_fit(figsize=(20,8))
```

```python
model_a.plot_predict_is(h=60, figsize=(20,8))
```

```python
model_a.plot_predict(h=60,past_values=100,figsize=(20,8))
```
