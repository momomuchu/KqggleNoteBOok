```python
df_avocado["Hash_Others_Volume"] = df_avocado["4046"] + df_avocado["4225"] + df_avocado["4770"] + df_avocado["Total Bags"]
df_price_volume = df_avocado[["AveragePrice", "Total Volume", "4046", "4225", "4770", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags", "year", "Hash_Others_Volume"]]
df_price_volume
```

```python
# Hash Avocados  : Small -> 4 oz, Large -> 9 oz, Extra Large -> 12.5 oz
# Other Avocados : Small -> 5 oz, Large -> 10 oz, Extra Large -> 16 oz
df_price_volume["Weight_Average_(oz)"] = (df_price_volume["4046"] * 4 + df_price_volume["4225"] * 9 + \
                                         df_price_volume["4770"] * 12.5 + df_price_volume["Small Bags"] * 5 + \
                                         df_price_volume["Large Bags"] * 10 + df_price_volume["XLarge Bags"] * 16) / df_price_volume["Hash_Others_Volume"]
df_price_volume
```

```python
# Weight Average
weight_average = df_price_volume["Weight_Average_(oz)"].sum() / df_price_volume["Weight_Average_(oz)"].count()
weight_average
```

```python
# 2015 -> $2.72/kg, 2016 -> $3.13/kg, 2017 -> $3.82/kg
df_year_price = pd.DataFrame({"year": [2015, 2016, 2017], "cost_price_(kg)": [2.72, 3.13, 3.82]})

# 1 kg -> 35.274 oz
df_year_price["cost_price_(unit)"] = df_year_price["cost_price_(kg)"] * weight_average / 35.274

df_year_price
```

```python
# Cost Price Unit Estimation in 2018
X = df_year_price['year'].values.reshape(-1, 1)
y = df_year_price['cost_price_(kg)']

reg = LinearRegression().fit(X, y)

price_2018 = reg.predict([[2018]])
print("Estimated Cost Price in 2018:", price_2018[0])

weight_average_unit_2018 = price_2018[0] * weight_average / 35.274
df_year_price.loc[3, :] = [2018, price_2018[0], weight_average_unit_2018]

print("Weight Average Unit 2018:", weight_average_unit_2018)
df_year_price
```

```python
# Datetime Features
df_date = pd.DataFrame({"date": df_avocado["Date"].unique()})
df_date = df_date.sort_values(by="date").reset_index(drop=True)
df_date['date'] = pd.to_datetime(df_date['date'])
df_date['month'] = df_date['date'].dt.month
df_date['weekofyear'] = df_date['date'].dt.weekofyear
df_date['quarter'] = df_date['date'].dt.quarter

df_date
```

```python
df_avocado_used = df_avocado[["region", "type", "Date", "AveragePrice", "Hash_Others_Volume", "year"]]
df_avocado_used['Date'] = pd.to_datetime(df_avocado_used['Date'])
df_avocado_used = df_avocado_used.sort_values(by=["region", "type", "Date"], ascending=True).reset_index(drop=True)
df_avocado_used = df_avocado_used.merge(df_year_price, left_on="year", right_on="year")
df_avocado_used = df_avocado_used.merge(df_date, left_on="Date", right_on="date")
df_avocado_used = df_avocado_used.drop("Date", axis=1)
df_avocado_used = df_avocado_used[["region", "type", "date", "year", "month", "weekofyear", "quarter", "AveragePrice", "Hash_Others_Volume", "cost_price_(kg)", "cost_price_(unit)"]]
df_avocado_used = df_avocado_used.rename(columns={"Hash_Others_Volume": "Volume"})
df_avocado_used["Profit"] = (df_avocado_used["AveragePrice"] - df_avocado_used["cost_price_(unit)"]) * df_avocado_used["Volume"]
df_avocado_used
```

```python
df_train = df_avocado_used[df_avocado_used["year"] < 2018].reset_index(drop=True)
df_test = df_avocado_used[df_avocado_used["year"] >= 2018].reset_index(drop=True)

print("Train Dataset")
display(df_train)

print("Test Dataset")
display(df_test)
```

```python
train_features = df_train[["region", "type", "year", "month", "weekofyear", "quarter", "AveragePrice"]]
test_features = df_test[["region", "type", "year", "month", "weekofyear", "quarter", "AveragePrice"]]

train_labels = df_train["Volume"]
test_labels = df_test["Volume"]

display(train_features)
display(test_features)
```

```python
categorical_features = ["region", "type"]

df_category = pd.get_dummies(data=train_features[categorical_features])
train_features = pd.concat([train_features, df_category], axis=1)
train_features = train_features.drop(categorical_features, axis=1)

df_category = pd.get_dummies(data=test_features[categorical_features])
test_features = pd.concat([test_features, df_category], axis=1)
test_features = test_features.drop(categorical_features, axis=1)

display(train_features)
display(test_features)
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
train_features_poly = poly.fit_transform(train_features)
test_features_poly = poly.transform(test_features)

lr_poly_model = Lasso()
lr_poly_model.fit(train_features_poly, train_labels)

test_pred = lr_poly_model.predict(test_features_poly)

n = test_features_poly.shape[0]
p = test_features_poly.shape[1]

mse = mean_squared_error(test_labels, test_pred)
rmse = mean_squared_error(test_labels, test_pred, squared=False)
mae = mean_absolute_error(test_labels, test_pred)
r2 = r2_score(test_labels, test_pred)
adj_r2 = adjusted_r2_score(test_labels, test_pred, n, p)

print("Metric Scores")
print("MSE    :", mse)
print("RMSE   :", rmse)
print("MAE    :", mae)
print("R2     :", r2)
print("Adj R2 :", adj_r2)
```

```python
# Column Name
feature_columns = test_features.columns.tolist()
poly_columns = feature_columns.copy()
for i in range(len(feature_columns)):
    for j in range(i+1, len(feature_columns)):
        poly_columns.append("{}_{}".format(feature_columns[i], feature_columns[j]))

train_features_poly = pd.DataFrame(train_features_poly)
train_features_poly.columns = poly_columns
test_features_poly = pd.DataFrame(test_features_poly)
test_features_poly.columns = poly_columns

test_features_poly
```

```python
test_results = pd.DataFrame({"Actual": test_labels, "Predicted": test_pred})
test_results
```

```python
list_region_r2_scores = []
for region in region_name:
    region_test_features = test_features_poly[test_features_poly["region_{}".format(region)]==1]
    index_region = region_test_features.index.tolist()
    region_test_labels = test_labels.loc[test_labels.index.isin(index_region)]
    region_test_preds = lr_poly_model.predict(region_test_features)
    
    region_r2 = r2_score(region_test_labels, region_test_preds)
    list_region_r2_scores.append(region_r2)
#     print(region)
#     print("R2 Scores:", region_r2)
    
df_region_r2 = pd.DataFrame()
df_region_r2["Region"] = region_name
df_region_r2["R2 Scores"] = list_region_r2_scores
df_region_r2 = df_region_r2.sort_values(by="R2 Scores", ascending=True).reset_index(drop=True)
df_region_r2
```

```python
def get_optimal_price(row):
    best_profits = [0]
    best_prices = -50
    list_demand = []
    list_profit = []

    cal_prices = np.arange(0.83, 3.51, 0.01)
    
    for pri in cal_prices:
        for i in range(len(feature_columns)):
            if(i<4):
                row["{}_AveragePrice".format(feature_columns[i])] = pri * row[feature_columns[i]]
            elif(i==4):
                row["AveragePrice"] = pri
            else:
                row["AveragePrice_{}".format(feature_columns[i])] = pri * row[feature_columns[i]]
                
        demand = lr_poly_model.predict([row])
        if(demand < 0):
            break
        list_demand.append(demand[0])
        profit = (pri - weight_average_unit_2018) * demand
        if(profit > best_profits):
            best_profits = profit
            best_prices = pri
        list_profit.append(profit[0])
    return best_profits[0], best_prices, list_demand, list_profit
```

```python
test_features_opt = test_features_poly.copy()

all_best = test_features_opt.apply(lambda x: get_optimal_price(x), axis=1)
test_features_opt["Best Profits"] = [x[0] for x in all_best]
test_features_opt["Optimal Price"] = [x[1] for x in all_best]
test_features_opt["List Demand"] = [x[2] for x in all_best]
test_features_opt["List Profit"] = [x[3] for x in all_best]

test_features_opt
```

```python
def date_range(year, week_of_year):
    first_date = datetime.datetime.strptime(f'{year}-W{week_of_year-1}-1', "%Y-W%W-%w").date()
    last_date = first_date + datetime.timedelta(days=6.9)
    
    first_date = first_date.strftime("%d/%m/%Y")
    last_date = last_date.strftime("%d/%m/%Y")
    return first_date, last_date

def optimal_price_plot(unit_region, unit_type, week_of_year):
    # Create DataFrame
    row_train = train_features_poly[
        (train_features_poly["region_{}".format("West")]==1) &
        (train_features_poly["type_{}".format("conventional")]==1)
    ]
    row_train_index = row_train.index.tolist()
    demand_train = train_labels[train_labels.index.isin(row_train_index)]
    
    row_test = test_features_opt[
        (test_features_opt["region_{}".format(unit_region)]==1) &
        (test_features_opt["type_{}".format(unit_type)]==1) &
        (test_features_opt["weekofyear"]==week_of_year)
    ].reset_index(drop=True).iloc[0, :]
    
    first_date, last_date = date_range(2018, int(row_test["weekofyear"]))
    cal_prices = np.arange(0.83, 3.51, 0.01)
    len_data = len(row_test["List Demand"])
    list_demand = row_test["List Demand"]
    list_profit = row_test["List Profit"]

    # Create Subplots
    fig = make_subplots(
        rows=2, cols=6,
        specs=[
            [{"type": "choropleth", "colspan": 2}, None, {"type": "indicator", "colspan": 2}, None, {"type": "indicator", "colspan": 2}, None],
            [{"type": "scatter", "colspan": 3}, None, None, {"type": "scatter", "colspan": 3}, None, None]
        ],
        subplot_titles=["", "", "", "Price vs Demand", "Price vs Profit"], 
        horizontal_spacing=0.10
    )

    # Region Map
    fig.add_trace(
        go.Choropleth(
            locations=df_region_code[unit_region], 
            z=[1] * df_region_code.shape[0],
            locationmode="USA-states",
            colorscale = [[0.0, "#b3c656"], [1.0, "#7b964a"]],
            hoverinfo="skip",
            showscale=False
        ), row=1, col=1
    )

    # Optimal Price Text
    fig.add_trace(
        go.Indicator(
            mode="number",
            value=row_test["Optimal Price"],
            number={"prefix": "$"},
            title={"text": "Optimal Price"}
        ), row=1, col=3
    )

    # Profit Text
    fig.add_trace(
        go.Indicator(
            mode="number",
            value=row_test["Best Profits"],
            number={"prefix": "$"},
            title={"text": "Maximum Profit"}
        ), row=1, col=5
    )

    # Price vs Demand
    fig.add_trace(
        go.Scatter(
            x=cal_prices[:len_data],
            y=list_demand,
            mode="lines",
            line=dict(
                color="#53783b",
                width=4
            ),
            name="Regression Line"
        ), row=2, col=1
    )
    
    fig.add_trace(
        go.Scatter(
            x=row_train["AveragePrice"],
            y=demand_train,
            mode="markers",
            marker=dict(
                color=row_train["year"],
                colorscale=[[0.0, "#fcdb7e"], [0.5, "#b3c656"], [1.0, "#7b964a"]],
                size=8,
            ),
            opacity=0.5,
            hovertemplate='''Price: <b>%{x}</b><br>Demand: <b>%{y}</b><br>Year: <b>%{marker.color}</b><br>''',
            name="Train Data"
        ), row=2, col=1
    )

    # Price vs Profit
    fig.add_trace(
        go.Scatter(
            x=cal_prices[:len_data],
            y=list_profit,
            mode="lines",
            line=dict(
                color="#b3c656",
                width=3.5
            ),
            name="Regression Line"
        ), row=2, col=4
    )
    
    fig.add_trace(
        go.Scatter(
            x=[row_test["Optimal Price"]],
            y=[row_test["Best Profits"]],
            mode="markers",
            marker=dict(
                size=14,
                color="#53783b"
            ),
            name="Optimal Price"
        ), row=2, col=4
    )

    fig.add_shape(
        type="line",
        x0=row_test["Optimal Price"], x1=row_test["Optimal Price"],
        y0=0.0, y1=row_test["Best Profits"],
        xref="x2", yref="y2",
        line=dict(
            color='#53783b',
            width=3,
            dash='dash'
        )
    )
    
    # Update Subplot Title
    fig.layout.annotations[0].update(y=0.57)
    fig.layout.annotations[1].update(y=0.57)
    
    # Update Axes
    fig.update_yaxes(domain=[0.6, 1.0], row=1, col=1)
    fig.update_yaxes(domain=[0.6, 1.0], row=1, col=3)
    fig.update_yaxes(domain=[0.6, 1.0], row=1, col=5)
    fig.update_xaxes(title="Price", showgrid=False, zeroline=False, linecolor="Gray", ticks="outside", row=2, col=1)
    fig.update_yaxes(title="Demand", showgrid=False, zeroline=False, linecolor="Gray", ticks="outside", domain=[0.0, 0.55], row=2, col=1)
    fig.update_xaxes(title="Price", showgrid=False, zeroline=False, linecolor="Gray", ticks="outside", row=2, col=4)
    fig.update_yaxes(title="Profit", showgrid=False, zeroline=False, linecolor="Gray", ticks="outside", domain=[0.0, 0.55], row=2, col=4)
    
    # Update Layout
    fig.update_layout(
        title='<b style="font-size:18px">Price Optimization</b><br><b style="font-size:13px">Region : {}</b><br><b style="font-size:13px">Type : {}</b><br><b style="font-size:13px">Date : {} - {}</b>'.format(unit_region, unit_type.capitalize(), first_date, last_date),
        title_x=0.5, title_y=0.97, font_family="Cambria",
        width=900, height=600,
        showlegend=False,
        plot_bgcolor="White",
        paper_bgcolor="White",
        geo=dict(
            scope="usa"
        ),
        margin_pad=3
    )

    # Show
    fig.show(renderer="iframe_connected")
```
